Table 2 - Number of prescription items1 and prescription items per 100,000 population, for the treatment of alcohol dependence, prescribed in primary care2 and dispensed3 in the community, by Commissioning Region and Clinical Commissioning Group

```python
from gssutils import *

if is_interactive():
    import requests
    from cachecontrol import CacheControl
    from cachecontrol.caches.file_cache import FileCache
    from cachecontrol.heuristics import LastModified
    from pathlib import Path

    session = CacheControl(requests.Session(),
                           cache=FileCache('.cache'),
                           heuristic=LastModified())

    sourceFolder = Path('in')
    sourceFolder.mkdir(exist_ok=True)

    inputURL = 'https://webarchive.nationalarchives.gov.uk/20180328130416/http://digital.nhs.uk/media/30889/Statistics-on-Alcohol-England-2017-Tables/Any/alc-eng-2017-tab'
    inputFile = sourceFolder / 'alc-eng-2017-tab'
    response = session.get(inputURL)
    with open(inputFile, 'wb') as f:
      f.write(response.content)    
```

```python
tab = loadxlstabs(inputFile, sheetids='Table 2')[0]
```

```python
observations = tab.excel_ref('F9').expand(DOWN).expand(RIGHT).is_not_blank()
observations
```

```python
Geography = tab.excel_ref('A9').expand(DOWN).is_not_blank()
Geography
```

```python
gr = tab.excel_ref('C13').expand(DOWN).is_not_blank()
gr
```

```python
treatment = tab.excel_ref('F7').expand(RIGHT).is_not_blank() 
treatment
```

```python
category = tab.excel_ref('F6').expand(RIGHT).is_not_blank() 
category
```

```python
Dimensions = [
            HDim(category,'Category',CLOSEST,LEFT),
            HDim(treatment,'Treatment of alcohol dependence',DIRECTLY,ABOVE),
            HDim(gr,'Commissioning Region and Clinical Commissioning Group',DIRECTLY,LEFT),
            HDim(Geography,'Geography',DIRECTLY,LEFT),
            HDimConst('Period','2016'),
            HDimConst('Unit','Prescription'),
            HDimConst('Measure Type','Count')
            ]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
# if is_interactive():
#     savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table.columns = ['Value' if x=='OBS' else x for x in new_table.columns]
```

```python
new_table.head()
```

```python
new_table['Treatment of alcohol dependence'] = new_table['Treatment of alcohol dependence'].map(
    lambda x: {
        'Total' : 'All' 
        }.get(x, x))
```

```python
new_table.head()
```

```python
new_table.tail()
```

```python
new_table.dtypes
```

```python
new_table['Value'] = new_table['Value'].astype(int)
```

```python
new_table['Period'] = new_table['Period'].astype(int)
```

```python
new_table.head(3)
```

```python
new_table['Category'] = new_table['Category'].str.rstrip('6, 7')
```

```python
new_table['Commissioning Region and Clinical Commissioning Group'].fillna('Unknown', inplace = True)
```

```python
new_table['Geography'].fillna('E92000001', inplace = True)
```

```python
new_table = new_table[['Geography','Period','Category','Treatment of alcohol dependence','Commissioning Region and Clinical Commissioning Group','Measure Type','Value','Unit']]
```

```python
if is_interactive():
    SubstancetinationFolder = Path('out')
    SubstancetinationFolder.mkdir(exist_ok=True, parents=True)
    new_table.to_csv(SubstancetinationFolder / ('table2.csv'), index = False)
```
